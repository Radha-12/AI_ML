Q2.An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions

A) Probability that none of the items are defective?

```python
import numpy             as np
import pandas            as pd
import scipy.stats       as stats
import matplotlib.pyplot as plt
```

```python
p   =  0.05 # failure rate for the product that the electrical manufacturing company is 5%
n   =  10 # sample size
k   =  np.arange(0,11) # An array of quantiles
```

```python
binomial = stats.binom.pmf(k,n,p)
```

```python
print(binomial)
```

```python
print('A) None of the items are defective? is %1.4f' %binomial[0])
```

Probability that none of the items are defective is 0.5987

B) Probability that exactly one of the items are defective? 

```python
print('B) None of the items are defective? is %1.4f' %binomial[1])
```

Probability that exactly one of the items are defective is 0.3151

C) Probability that two or fewer of the items are defective?

we need to calculate cumulative probability of Upto Two items being defective

```python
cumbinomial = stats.binom.cdf(k,n,p)
```

```python
print(cumbinomial)
```

```python
print('C) Two or fewer of the items are defective? is %1.4f' %cumbinomial[2])
```

Probablity of Two or fewer items are defective is 0.9885

D) Probability that three or more of the items are defective ?

we need to subtract cumulative Probability upto 2 defective items from 1.

```python
P = 1- cumbinomial[2]
print('D) Three or more of the items are defective is %1.4f' % P)
```

Probability of three or more of the items are defective is 0.0115

Q3. A car salesman sells on an average 3 cars per week.

A) What is Probability that in a given week he will sell some cars?

Here, 
 
 Î»(Average number of successes)  = 3

```python
import numpy             as np
import pandas            as pd
import scipy.stats       as stats
import matplotlib.pyplot as plt
```

```python
rate =  3 # Mean
n    =  np.arange(0,16)
cumpoisson = stats.poisson.cdf(n,rate)
```

```python
cumpoisson
```

```python
# P = probability that the salesman sells some cars
P   = 1 - cumpoisson[0]
print(' A) Probability that the salesman sells some cars is %1.4f' % P)
```

Probability that the salesman sells some cars is 0.9502

B) What is Probability that in a given week he will sell 2 or more but less than 5 cars? 

```python
# Let P1 be the probability that the salesman sells more than 4 cars
P1 = cumpoisson[4] # P(X >=4)
# Let P2 be the probability that the salesman sells more than 1 car
P2 = cumpoisson[1] # P(X >=1)
P = P1 - P2 # Prob. that the salesman will sell between 2 and 5 cars
print(' B) Probability that the salesman sells between 2 and 5 cars is %1.4f' % P)
```

Probabiliy that the salesman sells between 2 and 5 cars is 0.6161

C) Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

```python
poisson = stats.poisson.cdf(n,rate) # To plot a graph with poisson distribution function
plt.plot(poisson,n, 'o-') 
plt.title('Poisson') # To display the Title of the plot as 'Poisson'
plt.xlabel('Cumulative Prob of cars sold per week') # To display the xlabel as 'Cumulative Prob of cars sold per week'
plt.ylabel('Number of cars sold per week') # To display the ylabel as 'Number of cars sold per week'
plt.show() # To display the plot
```

* P(D) = probability that given test reports indicates how many are suffering from a disease (based on historical data)
* P(P|D) = probability that test reports shows positive which shows how many are suffering from a disease from the samples collected.
* P(ND) = probability that given test reports indicates how many are not suffering from a disease 
* P(P|ND) = probability that test reports shows negative which shows how many are not suffering from a disease from the samples collected.

From Bayes Theorem, 
      
      We know that

              P(D|P) = P(P|D)*P(D)/P(P)

      We also know that the test results positive or negative are the mutual exclusively events

              P(P) = P(P âˆ© D) + P(P âˆ© ND)
 
     which in turn can be written as 

              P(D|P) = P(P|D) * P(P) / P(P|D) * P(P) + P(P|ND) * P(ND)


 The data on the probabilities can be updated as we collect more and more samples in the above process. By setting a probability threshold, we can classify people who all are to be diagnonised.
```python
from datetime import date 
def Year(x):
  if 'to' in x:
    return x.split('to')[0]
  elif '-' in x:
    return x.split('-')[0]
  elif '~' in x:
    return x.split('~')[0]
  elif '_' in x:
    return x.split('_')[0]
  else:
    return x

df2["LaunchYear"] = df2["TeamLaunch"].apply(Year)
df2 = df2.astype({"LaunchYear": np.int64}) # Convert LaunchYear attribute data type to int
today = date.today()
# Create a new column with the name as " Team Age " which will define the age of the team in years
df2["Team Age"] = df2["LaunchYear"].apply(lambda x: today.year-x) # Subtract the current year with TeamLaunch year to get age of the team
df2.head() # To fetch first 5 rows of the data frame
```

*   **Win_P**= Percentage of 'WonGames' over 'PlayedGames'
*   **PFin**= Percentage of Podium finishes('TournamentChampion'+'Runner-up')over 'Tournament'
*   **SP**= 'Score' over 'PlayedGames' 
*   **TP**= Percentage of 'Tournament' over 'Team Age' 

```python
df2['Basket_D'] = (df2['BasketScored'] - df2['BasketGiven'])/df2['PlayedGames'].fillna(value=0)
df2['Win_P'] = df2['WonGames']/df2['PlayedGames'].fillna(value=0)*100
df2['PFin'] = df2['TournamentChampion'] + df2['Runner-up']/df2['Tournament']*100
df2['SP'] = df2['Score']/df2['PlayedGames'].fillna(value=0)
df2['TP'] = df2['Tournament']/df2['Team Age']*100
```

The distributions look very similar on the plot. However,to test the statistical significance, we will do statistical test using either Z-test or T-test as distributions are not normal, since our sample size > 30.

**Ho**=Null Hypothesis = There is no difference between two means(Mean of funds raised by Operating and Closed)

**Ha**= Alternate Hypothesis = There is a sigificant difference between two means(Mean of funds raised by Operating and Closed) 

Now, let us perform the Z-Test on this sample on this sample at 0.05 significance level.

```python
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest
alpha = 0.05 #defining the significance level
operating = df1[df1['OperatingState']=='Operating']['Funds_in_million']
closed = df1[df1['OperatingState']=='Closed']['Funds_in_million']
z_stat, p_val = ztest(operating, closed)

if p_val > alpha:
  print("Since p_val = %.3f is greater than the siginificant level (0.05), the difference between the means not significant and we fail to reject Ho" %p_val)
else:
  print("Since p_val = %.3f is less than the siginificant level (0.05), the difference between the means not significant" %p_val)  
```

```python
# Using T-test also gives the same result
from scipy.stats import ttest_ind
t_stat, p_val = ttest_ind(operating, closed)

if p_val > alpha:
  print("Since p_val = %.3f is greater than the siginificant level (0.05), the difference between the means not significant and we fail to reject Ho" %p_val)
else:
  print("Since p_val = %.3f is less than the siginificant level (0.05), the difference between the means not significant" %p_val) 
```

We see that the proportions are different but to determine if they are statiscally different. We will perform Z-test or Chi-square test. 

**Ho**= Null Hypothesis = The proportion of companies that are operating is the same in both winners and contestants

**Ha**= Alternate Hypothesis = The proportion of companies that are operating statistically different between winners and contestansts

Lets, perform Z-test on this sample at 0.05 significance level

```python
from statsmodels.stats.proportion import proportions_ztest

z_test, p_val = proportions_ztest([n_w_o,n_c_o],[n_winners,n_contestants])

if p_val > alpha:
  print("Since p_val = %.3f is greater than the siginificant level (0.05), the difference between the means are significant and we will fail to reject Ho" %p_val)
else:
  print("Since p_val = %.3f is greater than the siginificant level (0.05), the difference between the means are significant" %p_val)
```
